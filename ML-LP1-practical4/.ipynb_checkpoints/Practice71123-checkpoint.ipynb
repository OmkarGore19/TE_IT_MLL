```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("SMSSpamCollection",sep='\t',names=['label','text'])
```

```python
df
```

```python
import nltk
```

```python
nltk.download('stopwords')
```

```python
sent = "Hello  freinds ! ML is OP as well as ADBMS"
```

```python
from nltk.tokenize import word_tokenize

from nltk.corpus import stopwords
stwd = stopwords.words('english')

from nltk.stem import PorterStemmer
ps = PorterStemmer()

def clean_text(sent):
    tokens = word_tokenize(sent)
    clean = [word for word in tokens if word.isdigit() or word.isalpha()]
    clean = [ps.stem(word) for word in clean
             if word not in stwd]
    return clean
```

```python
clean_text(sent)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(analyzer = clean_text)
```

```python
x = df['text']
y = df['label']
```

```python
x_new = tfidf.fit_transform(x)
```

```python
x_new.shape
```

```python
sns.countplot(x=y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_new,y,test_size=0.25,random_state=1)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.naive_bayes import GaussianNB
model_nb = GaussianNB()
model_nb.fit(x_train.toarray(),y_train)
y_pred_nb = model_nb.predict(x_test.toarray())
```

```python
from sklearn.ensemble import RandomForestClassifier
model_rf = RandomForestClassifier(random_state=1)
model_rf.fit(x_train,y_train)
y_pred_rf = model_rf.predict(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
model_lr = LogisticRegression(random_state=1)
model_lr.fit(x_train,y_train)
y_pred_lr = model_lr.predict(x_test)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay, accuracy_score
from sklearn.metrics import classification_report
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_pred_nb)
plt.title('Naive bayes')
plt.show()
print(f" Accuracy is {accuracy_score(y_test,y_pred_nb)}")
print(classification_report(y_test,y_pred_nb))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_pred_rf)
plt.title('Random Forest')
plt.show()
print(f" Accuracy is {accuracy_score(y_test,y_pred_rf)}")
print(classification_report(y_test,y_pred_rf))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_pred_lr)
plt.title('LR')
plt.show()
print(f" Accuracy is {accuracy_score(y_test,y_pred_lr)}")
print(classification_report(y_test,y_pred_lr))
```

```python
#Hyper parameter tunning
from sklearn.model_selection import GridSearchCV
```

```python
params = {
    'criterion':['gini','entropy'],
    'random_state' :[0,1,2,3,4],
    'class_weight' : ['balanced','balanced_subsample']
}
```

```python
grid = GridSearchCV(model_rf, param_grid=params,cv= 5, scoring ='accuracy' )
```

```python
grid.fit(x_train,y_train)
```

```python
model_rf_hp = grid.best_estimator_
```

```python
y_pred_hp = model_rf_hp.predict(x_test)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_pred_hp)
plt.title('Gride Search')
plt.show()
print(f" Accuracy is {accuracy_score(y_test,y_pred_hp)}")
print(classification_report(y_test,y_pred_hp))
```
