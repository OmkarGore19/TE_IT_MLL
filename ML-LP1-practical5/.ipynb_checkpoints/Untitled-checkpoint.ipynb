```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("Mall_Customers.csv")
```

```python
df
```

```python
df.dtypes
```

```python
x = df[['Annual Income (k$)','Spending Score (1-100)']]
```

```python
x
```

```python
plt.title("Unclustered Data")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.scatter(x["Annual Income (k$)"],x["Spending Score (1-100)"])
```

```python
from sklearn.cluster import KMeans, AgglomerativeClustering
```

```python
sse =[]
for k in range(1,16):
    km = KMeans(n_clusters=k)
    km.fit_predict(x)
    sse.append(km.inertia_)
```

```python
plt.plot(range(1,16),y = sse)
plt.xlabel('Cluster')
plt.ylabel('SSE')
```

```python
#Method second or alternative for elbow method
from sklearn.metrics import silhouette_score
silh =[]
for k in range(2,16):
km = KMeans(n_clusters=k)
labels = km.fit_predict(x)
score = silhouette_score(x, labels)
silh.append(score)
```

```python
silh
```

```python
plt.bar(range(2,16),y = silh)
plt.xlabel('Cluster')
plt.ylabel('silh')
```

```python
km = KMeans(n_clusters=5,random_state=1)
```

```python
km.labels_
```

```python
cent = km.cluster_centers_
```

```python
plt.title('Clustered Data')
plt.scatter(x=x['Annual Income (k$)'],y=x['Spending Score (1-100)'],c=labels )
plt.scatter(cent[:,0],cent[:,1], s=200, color='red')
```

```python
agl = AgglomerativeClustering(n_clusters=5)
```

```python
alabels = agl.fit_predict(x)
```

```python
alabels
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.title('Agglomerative')
plt.scatter(x=x['Annual Income (k$)'],y=x['Spending Score (1-100)'], c= alabels)

plt.subplot(1,2,2)
plt.title('KMEANS')
plt.scatter(x=x['Annual Income (k$)'],y=x['Spending Score (1-100)'],c=labels )
plt.scatter(cent[:,0],cent[:,1], s=200, color='red')
```
