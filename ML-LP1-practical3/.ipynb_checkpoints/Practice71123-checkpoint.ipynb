```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Admission_Predict_Ver1.1.csv")
```

```python
df
```

```python
df = df.drop('Serial No.',axis=1)#no need
```

```python
df
```

```python
from sklearn.preprocessing import Binarizer
br = Binarizer(threshold = 0.75)
df['Chance of Admit '] = br.fit_transform(df[['Chance of Admit ']])
```

```python
df
```

```python
x = df[['GRE Score', 'TOEFL Score', 'University Rating', 'SOP', 'LOR ', 'CGPA','Research']]
y = df['Chance of Admit ']
y = y.astype('int')#float to int
```

```python
sns.countplot(x=y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=1)
```

```python
x_train.shape, x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.tree import DecisionTreeRegressor
model_dt = DecisionTreeRegressor(random_state=1)
model_dt.fit(x_train,y_train)
y_pred_dt = model_dt.predict(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
model_lr = LogisticRegression(random_state=1,max_iter=1000)
model_lr.fit(x_train,y_train)
y_pred_lr = model_lr.predict(x_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
model_rf = RandomForestRegressor(random_state=1)
model_rf.fit(x_train,y_train)
y_pred_rf = model_rf.predict(x_test)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import ConfusionMatrixDisplay, accuracy_score
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_pred_dt)
plt.title("CM for DT")
plt.show()
print(f"Accuracy is {accuracy_score(y_test,y_pred_dt)}")
print(classification_report(y_test,y_pred_dt))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_pred_lr)
plt.title("CM for LR")
plt.show()
print(f"Accuracy is {accuracy_score(y_test,y_pred_lr)}")
print(classification_report(y_test,y_pred_lr))
```
